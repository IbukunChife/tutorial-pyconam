```python
import pandas as pd
from sklearn.feature_extraction.text \
    import CountVectorizer
```

```python
dados = pd.read_json('alura.json')
```

```python
dados.head()
```

```python
count_vector = CountVectorizer()
```

```python
data_transform = count_vector.fit_transform(
    dados['text']
)
```

```python
data_transform = pd.DataFrame(
    data_transform.toarray(),
    columns=count_vector.get_feature_names()
)
```

```python
data_transform
```

```python
for column in data_transform.columns.values:
    if column.isdigit():
        del data_transform[column]
```

```python
len(data_transform.columns)
```

```python
data_transform.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste,y_treino, y_teste = \
    train_test_split(
        data_transform, dados['category']
    )
```

```python
y_treino.head()
```

```python
model = MultinomialNB()
model.fit(x_treino, y_treino)
```

```python
model.score(x_teste, y_teste)
```
